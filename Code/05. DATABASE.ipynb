### 0. DataFrame Setting

### 1. CAD, CHF, EUR, JPY(100), USD

```python
# 환율 데이터 로딩
import pandas as pd
foreign_currency = pd.read_csv('../data/DAILY_DATA.csv')

# 필요한 국가만 추출
Dataset = foreign_currency[['date', 'CAD', 'CHF', 'EUR', 'GBP', 'JPY(100)', 'USD']]

# Index를 날짜로 변경
Dataset = Dataset.set_index('date')
```

### 2. Bitcoin

```python
# 경고 메세지 안뜨게 하기
import warnings
warnings.filterwarnings(action='ignore') 

# 데이터 불러오기
import pandas as pd
bitcoin = pd.read_csv('../data/Bitcoin.csv')

# 날짜 형식 변경하기
date = bitcoin['날짜'].apply(lambda x : x.replace('년 ','-'))\
                     .apply(lambda x : x.replace('월 ','-'))\
                     .apply(lambda x : x.replace('일',''))
bitcoin['날짜'] = date

# 칼럼명 변경하기
bitcoin.columns = ['date', 'BITCOIN', 'open', 'high', 'low', 'volume', 'rate']

# 날짜 인덱스로 변경하기
bitcoin = bitcoin.set_index('date')

# 비트코인 값 양식 바꾸기
value = bitcoin['BITCOIN'].apply(lambda x : x.replace(',','')).apply(lambda x : float(x))

# 데이터 프레임을 변경하기
bitcoin = pd.DataFrame(value)

# 2018년 2월 데이터부터 추출하고, 기존 데이터와 merge
Dataset = pd.concat([Dataset, bitcoin], axis=1)['2018-02-01':]
Dataset = Dataset.dropna()
```

### 3. Tether

```python
# 데이터 불러오기
import pandas as pd
tether = pd.read_csv('../data/tether.csv')

# 날짜에 필요없는 부분 제거
date = tether['snapped_at'].apply(lambda x : x.replace(' 00:00:00 UTC',''))
tether['date'] = date

# 날짜를 인덱스로 바꾸기
tether = tether.set_index('date')

# 칼럼명 지정하기
tether.columns = ['time', 'TETHER', 'market_cap', 'total_volume']

# 테더만 빼오기
tether = tether['TETHER']

# 기존 데이터와 merge
Dataset = pd.concat([Dataset, tether], axis=1)['2018-02-01':]
Dataset = Dataset.dropna()
```
